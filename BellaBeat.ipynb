## **Data Manipulation**


#### The following Data Manipulation will be performed according my recognization.

1. Converting "ActivityDate" column's datatype into datetime.
2. Convert datetime format like yyyy-mm-dd.
3. I will create a new column to separate the weekdays and the new column name will be "DayOfTheWeek".
4. will create new column "TotalMins" which will be the sum of 'VeryActiveMinutes', 'FairlyActiveMinutes',
'LightlyActiveMinutes', 'SedentaryMinutes'.
5. Create new column "TotalHours" by converting minutes which related to #4 from minutes to hours.
6. Rearrange and rename columns for better understanding.


```python
# convert "ActivityDate" col's datatype into "dtaetime" adn format will be "yyyy-mm-dd":-

daily_activity["ActivityDate"] = pd.to_datetime(daily_activity["ActivityDate"], format="%m/%d/%Y")

# Re-pint info to check datatype:-
daily_activity.info()
```

Yes. it's working perfectly.. we have changed the datatype of "ActivityDate" into DateTime.

* let's check "ActivityDate" column..

```python
daily_activity["ActivityDate"].head()
```

##### Now, we are creating some new columns, rename columns name rearrange columns etc.
* I will create new columns like "DayOfTheWeek", "TotalExerciseMinutes", "TotalExerciseHours".
* and rename dataset from "daily_activity" to "df_activity".

```python
# create new list of rearrange the columns from dataset:-

new_col = ["Id", "ActivityDate", "DayOfTheWeek", 'TotalSteps', 'TotalDistance', 'TrackerDistance',
       'LoggedActivitiesDistance', 'VeryActiveDistance',
       'ModeratelyActiveDistance', 'LightActiveDistance',
       'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes',
       'LightlyActiveMinutes', 'SedentaryMinutes',"TotalExerciseMinutes", "TotalExerciseHours", 'Calories']

# "reindex" function to rearrange columns based on "new_col"

df_activity = daily_activity.reindex(columns=new_col)

df_activity.head()
```

```python
# create new column by seprating the date into "DayOfTheWeek":-

df_activity["DayOfTheWeek"] = df_activity['ActivityDate'].dt.day_name()
df_activity["DayOfTheWeek"].head()
```

```python
# rearranging the column names for better undersatnding:-

df_activity.rename(columns = {"Id":"ID", "ActivityDate": "Date", "DayOfTheWeek":"Days", 'TotalSteps':"Total_steps",
                              'TotalDistance': "Total_Dist", 'TrackerDistance': "Track_dist",
                              'LoggedActivitiesDistance': "Logged_dist", 'VeryActiveDistance': "Very_active_dist",
                              'ModeratelyActiveDistance': "Moderate_active_dist", 'LightActiveDistance': "Light_active_dist",
                              'SedentaryActiveDistance': "Sedentary_active_dist", 'VeryActiveMinutes': "Very_active_Min",
                              'FairlyActiveMinutes': "Fair_active_Min",
                              'LightlyActiveMinutes': "Light_active_Min", 'SedentaryMinutes': "Sedentary_active_Min",
                              "TotalExerciseMinutes": "Total_Mins", "TotalExerciseHours": "Total_Hours",
                              'Calories':"Calor" }, inplace=True)

df_activity.columns

```

```python
df_activity.head()
```

###### We created Total_mins column. So for that I will do sum of total time logged.

```python
df_activity["Total_Mins"] = df_activity["Very_active_Min"] + df_activity["Fair_active_Min"] + df_activity["Light_active_Min"] + df_activity["Sedentary_active_Min"]
df_activity["Total_Mins"].head()
```

##### One more column we already created, which was "Total_Hours"....
So, here we will convert our Total_Mins into hours.

```python
# Colculation for converting Total minutes into total hours.

df_activity["Total_Hours"] = round(df_activity["Total_Mins"] / 60)

df_activity["Total_Hours"].head()
```

```python
# Let's check our whole data frame again...

df_activity.head()
```

##### So, as we did the statistic calculation for total minutes and convert it into hours so both columns values are shows up in the dataframe..

#### *Data VIasualization and Findings:-*


My first Visualization will be to check users' frequency of using the app in a week. for that I will muse "Days" column from the data frame.

```python
# import matplotlib package:-

import matplotlib.pyplot as plt

# Plotting Histogram :-

plt.style.use("default")
plt.figure(figsize=(10,8))
plt.hist(df_activity.Days, bins = 7,
         width = 0.5, color = "purple", edgecolor = "red")

# adding annotation and visuals:-

plt.xlabel("Days of the Week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app across the week")
plt.grid(True)
plt.show()
```

#### Conclusion:-

In this hostogram plot, we are looking the frequency of FitBit app usage in terms of the week.

* as I can see in the plot I discovered users prefer to track their activity on the app during midweek from Tuesday to Friday which menas only 4 days in a week.
* Users' frequency dropped down from Friday to Monday. So, we can say that most users are not preferring to use the app on weekends.

```python
# creating scatter plot:-

plt.style.use("bmh")
plt.figure(figsize = (10,8))
plt.scatter(df_activity.Total_steps, df_activity.Calor,
            alpha = 0.8, c = df_activity.Calor,
            cmap = "Spectral")

# adding annotations and visuals:-

median_calories = 2303
median_steps = 7637

plt.colorbar(orientation = "vertical")
plt.axvline(median_steps, color = "red", label = "Median steps")
plt.axhline(median_calories, color = "purple", label = "Median Calories Burned")
plt.xlabel("steps taken")
plt.ylabel("Calories burned")
plt.title("Calories burned for every steps taken")
plt.grid(True)
plt.legend()
plt.show()
```

* This is a positive correlation between calories burned and how many steps taken by users.
* That means if more steps are taken by users more calories are burned.
* As we can see in the plot points are lying in between 5000 to 10000 steps and for that, 2000 to 3000 calories can be buned.
* As go beyond 15000 steps calories will burned more than 3000.
* here, there are some outliers like 15000 steps was there but calories burned only 1500. which is not appropriate. So there are some exceptions.


```python
# creating a scatter plot for average steps, sedentary hours and average hours.

plt.style.use("grayscale")
plt.figure(figsize = (10,8))
plt.scatter(df_activity.Total_Hours, df_activity.Calor,
            alpha = 0.8, c = df_activity.Calor,
            cmap = "Spectral")

# Adding annotation and visuals:-
median_calories = 2303
median_hours = 20
median_sedentary_hur = 991 / 60

plt.colorbar(orientation = "vertical")
plt.axvline(median_hours, color = "Blue", label = "Median Steps")
plt.axvline(median_sedentary_hur, color = "Red", label = "Median Sedentary hours")
plt.axhline(median_calories, color = "Purple", label = "Medeian Hours")
plt.xlabel("Hours logged")
plt.ylabel("Calories burned")
plt.title("Calories burned for every hour logged")
plt.legend()
plt.grid(True)
plt.show()
```

let's check very active minutes, fair active minutes, and lightly active minutes by creating pie chart.

```python
# creating Pie Chart:-
# calculating total of individual minutes column:-

Very_active_Min = df_activity["Very_active_Min"].sum()
Fair_active_Min = df_activity["Fair_active_Min"].sum()
Light_active_Min = df_activity["Light_active_Min"].sum()
Sedentary_active_Min = df_activity["Sedentary_active_Min"].sum()

# plotting pie chart:-
slices = [Very_active_Min, Fair_active_Min, Light_active_Min, Sedentary_active_Min]
labels = [" Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary active minutes"]
color = ["Red", "Orange", "Yellow", "lightcoral", "Blue"]
explode = [0, 0, 0, 0.1]
plt.style.use("seaborn")
plt.pie(slices, labels=labels,
        colors = color, wedgeprops = {"edgecolor":"black"},
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```

### Conclusion:-

* Here, Sedentary minutes are taking the biggest slice which is 81.3%.
* This idicates that users are using the FitBit app to log daily activities such as daily commutes, inactive movements(moving from one spot to another) or running errands.
* App is rarely being used to track fitness as per the minor percentage of fairly active activity(1.1%) and very active activity(1.7%). This is highly discourageing a FitBit app was developed to encourage fitness.
